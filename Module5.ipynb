```python
import matplotlib.pyplot as plt
```

```python
savings = [0,100,150,200,300,500,300,350,500,600,750,100]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
#why do we need plt.show()?  It shows with just plt.plot(savings)
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savngs ($)')
plt.title('Saving Account Results')
plt.show()
```
